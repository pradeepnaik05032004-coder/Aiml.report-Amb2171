```python
import random
#simulating random predictions
for i in range(10):
  prediction=random.choice(["spam","not spam"])
  print(f"email.{i+1}:{prediction}")
```

```python
import random
#simulate 10 dice rolls
sample=[random.randint(1,6) for _ in range(10)]
print("sample rolls",sample)
print("sample mean:",sum(sample)/len(sample))
```
